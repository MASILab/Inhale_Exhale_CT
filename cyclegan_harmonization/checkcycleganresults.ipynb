```python
import os 
import nibabel as nib 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
validation_data_path = "/nfs/masi/krishar1/SPIE_2025_InhaleExhaleCT/data_split/valid/inspiratory_BONE"
validation_data_list = sorted(os.listdir(validation_data_path))

#Make a 5 by 5 subplot of the middle slices for image in the validation data 
fig, ax = plt.subplots(5,5, figsize=(20,20))
for i in range(5):
    for j in range(5):
        img = nib.load(os.path.join(validation_data_path, validation_data_list[i*5+j]))
        img_data = img.get_fdata()
        ax[i,j].imshow(np.rot90(img_data[:,:,img_data.shape[2]//2]), cmap='gray', vmin = -150, vmax = 150)
        ax[i,j].axis('off')
```
