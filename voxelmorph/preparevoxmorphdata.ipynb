```python
import os 
import pandas as pd 
import nibabel as nib
import numpy as np
```

```python
test_data_path = "/nfs/masi/krishar1/SPIE_2025_InhaleExhaleCT/data_split/test_resampled"
bone = "inspiratory_BONE"
standard = "expiratory_STANDARD"

bone_data = os.path.join(test_data_path, bone)
standard_data = os.path.join(test_data_path, standard)

bone_files = os.listdir(bone_data)
standard_files = os.listdir(standard_data)

bone_files
```

```python
bone_images = {}
standard_images = {}
for file in bone_files:
    name = file.split("_BONE")[0]
    print(name)
    bone_images[name] = os.path.join(bone_data, file)
for file in standard_files:
    name = file.split("_STANDARD")[0]
    standard_images[name] = os.path.join(standard_data, file)


bone_images, standard_images
```

```python
#Create a dataframe to store the data
#Sort the dictionary by keys
bone_images = dict(sorted(bone_images.items()))
standard_images = dict(sorted(standard_images.items()))

df = pd.DataFrame(columns = ["FileName", "inspiratory_BONE", "expiratory_STANDARD"])
df["FileName"] = bone_images.keys()
df["inspiratory_BONE"] = bone_images.values()
df["expiratory_STANDARD"] = standard_images.values()
```

```python
df.to_csv("/nfs/masi/krishar1/SPIE_2025_InhaleExhaleCT/data_split/VoxelMorph_data_inhale_exhale_CT.csv", index = False)
```
