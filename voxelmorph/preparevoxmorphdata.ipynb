```python
test_data_path_fixed = "/nfs/masi/krishar1/SPIE_2025_InhaleExhaleCT/data_split/registration_data/non_harmonized/resampled/resampled_test_masked_out_BONE"
test_data_path_moving = "/nfs/masi/krishar1/SPIE_2025_InhaleExhaleCT/data_split/registration_data/non_harmonized/resampled/resampled_test_masked_out_rigid_registered_STANDARD"

bone_files = os.listdir(test_data_path_fixed)
standard_files = os.listdir(test_data_path_moving)

bone_files, standard_files
```

```python
#Create a dataframe to store the data
#Sort the dictionary by keys
bone_images = dict(sorted(bone_images.items()))
standard_images = dict(sorted(standard_images.items()))

df = pd.DataFrame(columns = ["FileName", "inspiratory_BONE", "expiratory_STANDARD"])
df["FileName"] = bone_images.keys()
df["inspiratory_BONE"] = bone_images.values()
df["expiratory_STANDARD"] = standard_images.values()
```
